```python
import pandas as pd

# Load the dataset
df = pd.read_csv('final_food_delivery_dataset.csv')

# Q1: Gold members
print(f"Gold Orders: {df[df['membership'] == 'Gold'].shape[0]}")

# Q2: Hyderabad Revenue
print(f"Hyderabad Revenue: {round(df[df['city'] == 'Hyderabad']['total_amount'].sum())}")

# Q3: Distinct Users
print(f"Distinct Users: {df['user_id'].nunique()}")

# Q4: Average Order Value (Gold)
print(f"Gold AOV: {round(df[df['membership'] == 'Gold']['total_amount'].mean(), 2)}")

# Q5: Rating >= 4.5
print(f"High Rated Orders: {df[df['rating'] >= 4.5].shape[0]}")

# Q6: Top Gold City
gold_df = df[df['membership'] == 'Gold']
top_city = gold_df.groupby('city')['total_amount'].sum().idxmax()
print(f"Orders in {top_city}: {gold_df[gold_df['city'] == top_city].shape[0]}")
```
